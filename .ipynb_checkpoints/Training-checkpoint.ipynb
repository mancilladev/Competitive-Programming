```python
import sqlite3
import numpy as np
import pandas as pd
import rpy2.common as com
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 6)
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('codeforces/training.csv')
df.tail()
```

```python
df_problems = pd.read_csv('~/codeforces-api/ui/problem_ratings.csv')
df_problems.tail()
```

```python
rating = dict()
for index, row in df_problems.iterrows():
    problem = str(row['contestID']) + row['problemID']
    rating[problem] = row['problemRating']
```

```python
for index, row in df.iterrows():
    problem = row['problem']
    df.loc[index, 'problemRating'] = rating[problem]
df.tail()
```

## Brief analysis

```python
df.loc[:, ['ac','seconds','problemRating']].plot(subplots=True, kind='density', sharex=False)
```

```python
df_ac = df['ac'].value_counts(sort=True).to_frame()
df_ac.plot.barh(color='green')
```

```python
df_day = df['day'].value_counts(sort=False).to_frame()
df_day.plot(color='purple')
```

## Comparisons

```python
df_corr = df.loc[:, ['ac','seconds','problemRating']].corr()
plt.matshow(df_corr)
plt.xticks(range(len(df_corr.columns)), df_corr.columns)
plt.yticks(range(len(df_corr.columns)), df_corr.columns)
plt.colorbar()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_ylabel('AC', color='green')
ax2 = ax1.twinx()
ax2.set_ylabel('Seconds', color='blue')

df.plot(x=df.index, y='ac', color='green', ax=ax1)
df.plot(x=df.index, y='seconds', color='blue', ax=ax2, grid=False)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_ylabel('AC', color='green')
ax2 = ax1.twinx()
ax2.set_ylabel('Problem Rating', color='magenta')

df.plot(x=df.index, y='ac', color='green', ax=ax1)
df.plot(x=df.index, y='problemRating', color='magenta', ax=ax2, grid=False)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_ylabel('Seconds', color='blue')
ax2 = ax1.twinx()
ax2.set_ylabel('Problem Rating', color='magenta')

df.plot(x=df.index, y='seconds', color='blue', ax=ax1)
df.plot(x=df.index, y='problemRating', color='magenta', ax=ax2, grid=False)
```

# Rolling

```python
rolling = df.rolling(window=3).mean()
```

## AC

```python
ax_df = df.plot(y='ac', color='g', lw=2)
rolling.plot(y='ac', color='black', style='--', ax=ax_df, legend=0)
```

```python
sns.regplot(x='index', y='ac', data=df.reset_index(), color='green')
```

## Seconds

```python
ax_df = df.plot(y='seconds', color='b', lw=2)
rolling.plot(y='seconds', color='black', style='--', ax=ax_df, legend=0)
```

```python
sns.regplot(x='index', y='seconds', data=df.reset_index(), color='blue')
```
